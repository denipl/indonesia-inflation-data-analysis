```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data = pd.read_csv('data.csv',header=3)
data=data.rename(columns={'Unnamed: 0':'Kota'})
data['Januari'] = data['Januari'].str.replace('-', '')
data=data.replace('-', "",regex=True)
data=data.drop(columns=['Tahunan'])
bulan = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
data[bulan] = data[bulan].apply(pd.to_numeric, errors='coerce')
data.head()
```

```python
mom = data[bulan].pct_change(axis=1)* 100
mom.head()
```

```python
plt.figure(figsize=(16,8))

sns.heatmap(
    data[bulan],
    cmap="coolwarm",
    annot=False
)

plt.title("Heatmap Inflasi Bulanan per Kota")
plt.xlabel("Bulan")
plt.ylabel("Kota")

plt.show()
```
